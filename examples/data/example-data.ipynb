Target needs to be numerical for most sklrean mtds (as of writing).

Also this subsets the data so that the pipeline can be demoed.

Finally, `data.csv` is assumed to be the adult dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("adult_subset.csv")
```

```python
data["Income"] = data["Income"].apply(lambda x: 1 if x == "<=50K" else 0)
```

```python
# Randomly subset -- 250 rows.
# assume the target column is "income" ("<=50K" / ">50K")
df_small, _ = train_test_split(
    data,
    train_size=20,
    stratify=data["Income"],  # preserve label distribution
    random_state=42,
)


df_small.to_csv("adult_subset.csv")
```
